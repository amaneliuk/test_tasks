```python
export_400_m = pd.read_csv('export_400_m.csv', sep = ';', encoding = 'ANSI')
export_400_m.head(5)
```

```python
vice_fp = export_400_m[export_400_m['Структурный элемент (результат ФП или мероприятие ГП)'] == 'Результат ФП']\
    .groupby('Заместители Председателя Правительства')['Наименование мероприятия (результата)'].count()\
    .sort_values(ascending = False)\
    .to_frame()
```

```python
vice_gp = export_400_m[export_400_m['Структурный элемент (результат ФП или мероприятие ГП)'] == 'Мероприятие ГП']\
    .groupby('Заместители Председателя Правительства')['Наименование мероприятия (результата)'].count()\
    .sort_values(ascending = False)\
    .to_frame()
```

```python
total_vice = vice_fp.merge(vice_gp, how = 'left', on = 'Заместители Председателя Правительства', suffixes=('_ФП', '_ГП'))\
    .fillna(0)\
    .astype('int')
```

```python
total_vice['Итого'] = total_vice['Наименование мероприятия (результата)_ФП']\
    + total_vice['Наименование мероприятия (результата)_ГП']
```

```python
total_vice = total_vice.sort_values(by = 'Итого', ascending = False)
total_vice
```

```python
total_vice.to_excel('total_vice.xlsx')
```

```python
total_vice\
    .sort_values(by = 'Итого', ascending = True)\
    .plot(kind = 'barh', figsize = (8,10))


plt.show()
```

```python
foiv_fp = export_400_m[export_400_m['Структурный элемент (результат ФП или мероприятие ГП)'] == 'Результат ФП']\
    .groupby('Организация')['Наименование мероприятия (результата)'].count()\
    .sort_values(ascending = False)\
    .to_frame()

foiv_gp = export_400_m[export_400_m['Структурный элемент (результат ФП или мероприятие ГП)'] == 'Мероприятие ГП']\
    .groupby('Организация')['Наименование мероприятия (результата)'].count()\
    .sort_values(ascending = False)\
    .to_frame()
```

```python
total_foiv = foiv_fp.merge(foiv_gp, how = 'outer', on = 'Организация', suffixes=('_ФП', '_ГП'))\
    .fillna(0)\
    .astype('int')
```

```python
total_foiv['Итого'] = total_foiv['Наименование мероприятия (результата)_ФП']\
    + total_foiv['Наименование мероприятия (результата)_ГП']
```

```python
total_foiv = total_foiv.sort_values(by = 'Итого', ascending = False)
total_foiv
```

```python
total_foiv.to_excel('total_foiv.xlsx')
```

```python
f_data = export_400_m.merge(data_4841, how = 'left', left_on = 'global_id', right_on = 'global_object_id')
```

```python
f_data.drop(['system_object_id', 'Комментарий МинЭк', 'Комментарий АЦ', ' Позиция ведомства', 'global_object_id', \
           'Показатель достижения цели_y', 'Годовой компонент', 'Помесячный компонент'], axis = 1, inplace = True)
```

```python
f_data.head(1)
```

```python
fp_targets = f_data[f_data['Структурный элемент (результат ФП или мероприятие ГП)'] == 'Результат ФП']\
    .groupby('Показатель достижения цели_x')['Наименование мероприятия (результата)'].count()\
    .to_frame()\
    .sort_values(by = 'Наименование мероприятия (результата)', ascending = False)
```

```python
gp_targets = f_data[f_data['Структурный элемент (результат ФП или мероприятие ГП)'] == 'Мероприятие ГП']\
    .groupby('Показатель достижения цели_x')['Наименование мероприятия (результата)'].count()\
    .to_frame()\
    .sort_values(by = 'Наименование мероприятия (результата)', ascending = False)
```

```python
total_targets = fp_targets.merge(gp_targets, how = 'outer', on = 'Показатель достижения цели_x', suffixes=('_ФП', '_ГП'))\
    .fillna(0)\
    .astype('int')
```

```python
total_targets['Итого'] = total_targets['Наименование мероприятия (результата)_ФП']\
    + total_targets['Наименование мероприятия (результата)_ГП']
```

```python
total_targets = total_targets.sort_values(by = 'Итого', ascending = False)
total_targets
```
